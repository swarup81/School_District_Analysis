```python
# Dependencie
import pandas as pd
```

```python
# Load in file
movie_file = "Resources/movie_scores.csv"
```

```python
# Read and display the CSV with Pandas
movie_file_df = pd.read_csv(movie_file)
movie_file_df
```

```python
# List all the columns in the table
movie_file_df.columns
```

```python
# We only want IMDb data, so create a new table that takes the Film and all the columns relating to IMDB
imdb_df = movie_file_df[["IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
imdb_df 
```

```python
# We only like good movies, so find those that scored over 7, and ignore the norm rating
good_movies = movie_file_df.loc[movie_file_df["IMDB"]>7, ["IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
good_movies.head()
                                
```

```python
# Find less popular movies--i.e., those with fewer than 20K votes
lesspopular_movies = movie_file_df.loc[movie_file_df["IMDB_user_vote_count"]<20000, ["IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
lesspopular_movies.head()
```
